#### Matplotlib expects the color channels in this order -- RGB
#### cv2 gives the results in the -- BGR

#### Hence to see the image in correct format we need to convert the BGR format array into RGB format array

```python
pic_fix = cv2.cvtColor(pic_arr,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(pic_fix)
```

```python
image = cv2.imread('00-puppy.jpg')

while True:
    cv2.imshow('puppy_2', image)

    # Wait 1ms & check if the user pressed ESC key.
    if cv2.waitKey(1) & 0xFF == 27:
        break;

# Destroting all windows which were opened by opencv
cv2.destroyAllWindows()
    
```
