```python
import numpy as  np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('giraffes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image.shape
```

```python
## Applying the thresholding to convert the image into black & white
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
image.shape
```

```python
black_white_giraffe = cv2.threshold(image, thresh=110, maxval=255, type=cv2.THRESH_BINARY)
```

```python
plt.imshow(black_white_giraffe[1], cmap='gray')
```

```python
##Blurring the image using the filter2D
```

```python
## Creating the kernel
```

```python
kernel = np.ones((4,4), dtype=np.float32)*0.1
kernel
```

```python
blurred_image = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
plt.imshow(blurred_image)
```

```python
## Detecting the edge of the image by applying the Sobel Filter
```

```python
## Converting the RGB image into gray scale version so that we can apply the sobel filter
```

```python
image2 = cv2.imread('giraffes.jpg', 0)
# image_gray = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)
plt.imshow(image2, cmap='gray')
```

```python
## Horizontal Edge detetection
```

```python
sobelx = cv2.Sobel(image2, cv2.CV_64F, 1, 0, 5)
plt.imshow(sobelx, cmap='gray')
```

```python
## Plot the color histogram for red, blue, green channels of the giraffee image
```

```python
img =cv2.imread('giraffes.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    print(i, col)
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Giraffes Histograms')
plt.show()
```
