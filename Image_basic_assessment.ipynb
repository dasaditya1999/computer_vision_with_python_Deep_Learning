```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('dog_backpack.jpg')
plt.imshow(img)
```

```python
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
```

```python
plt.imshow(cv2.flip(rgb_image,0))
```

```python
plt.imshow(cv2.flip(rgb_image, -1))
```

```python
plt.imshow(cv2.flip(rgb_image,1))
```

```python
cv2.rectangle(rgb_image, pt1=(200,400),pt2=(600,800),thickness=5,color=(255,0,0))
```

```python
plt.imshow(rgb_image)
```

```python
cv2.line(rgb_image, pt1=(200,800),pt2=(400,400),thickness=5,color=(0,0,255))
plt.imshow(rgb_image)
```

```python
cv2.line(rgb_image, pt1=(200,800),pt2=(620,800),thickness=5,color=(0,0,255))
cv2.line(rgb_image, pt1 = (620,800), pt2 = (400,400), thickness=5, color=(0,0,255))
plt.imshow(rgb_image)
```

```python
points = np.array([[400,400],[200,800],[650,800]])
```

```python
points = points.reshape(-1,1,2)
```

```python
points
```

```python
cv2.polylines(rgb_image, pts=[points], isClosed = True, color=(5,255,9), thickness=6)
plt.imshow(rgb_image)
```

```python
cv2.polylines(rgb_image, pts=[points], isClosed = True, color=(0,255,255), thickness=10)
plt.imshow(rgb_image)
```
