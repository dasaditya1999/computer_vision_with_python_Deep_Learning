```python
blank_image = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_image.shape
```

```python
blank_image[:,:,0] = 255
blank_image[:,:,1] = 170
blank_image[:,:,2] = 120
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image, pt1=(100,100), pt2=(300, 300), color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image, pt1=(200,200),pt2=(300,300), color=(100,200,255), thickness=10)
```

```python
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image, radius=40, center = (200,400), thickness=5, color=(222,0,0))
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image, radius=40, center = (200,400), thickness=-1, color=(222,0,0))
plt.imshow(blank_image)
```

```python
cv2.line(img=blank_image, pt1=(0,500), pt2=(400,250), color=(255,0,0),thickness=10)
plt.imshow(blank_image)
```

```python
cv2.line(img=blank_image, pt1=(100,300), pt2=(400,250), color=(255,255,0),thickness=5)
plt.imshow(blank_image)
```

```python
cv2.rectangle(img=blank_image, pt1=(50,100),pt2=(300,300),color=(0,255,0),thickness=-1)
```

```python
plt.imshow(blank_image)
```
