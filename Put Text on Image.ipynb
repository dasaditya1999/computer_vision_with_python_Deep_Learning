```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pic_arr = np.ones(shape=(756,712,3), dtype=np.int8)
```

```python
plt.imshow(pic_arr)
```

```python
cv2.circle(img=pic_arr, center=(400,400), radius=5, color=(255,0,0), thickness=10)
plt.imshow(pic_arr)
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
```

```python
cv2.putText(img=pic_arr,text='HEY Google', org=(300,300), fontScale=5, fontFace=font, color=(255,255,0), thickness=5)
plt.imshow(pic_arr)
```

```python
cv2.putText(img=pic_arr, text='Hey Google', org=(100,300), fontScale=2, fontFace=font, color=(255,255,0), thickness=5)
plt.imshow(pic_arr)
```
